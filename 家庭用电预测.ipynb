```python
# 创建一个时间字符串格式化字符串
def date_format(dt):
    t=time.strftime(' '.join(dt),'%d/%m/%Y %H:%M:%S')
    return (t.tm_year,t.tm_mon,t.tm_mday,t.tm_hour,t.tm_min,t.tm_sec)
```
