```python
# 导入模块
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

#在行内显示
%matplotlib inline 
mpl.rcParams['font.sans-serif']=['SimHei']
mpl.rcParams['axes.unicode_minus']=False
```

```python
plt.plot([1,2,3],[5,7,4])
plt.show()
```

```python
x=np.arange(-3,3,0.1)
y1=np.sin(x)
y2=np.cos(x)
```

```python
plt.figure()#创建第一个图形
plt.plot(x,y1)
plt.figure(num=3,figsize=(6,5))#创建第二个图形
plt.plot(x,y2)
plt.show()
```

```python
plt.plot([1,2,3],[5,7,4],color='red',linestyle='dashed',marker='o'
         ,markersize=10,alpha=0.5)
plt.show()
```

```python
x1=[1,2,3]
y1=[5,7,4]
x2=[1,2,3]
y2=[10,14,12]
plt.plot(x1,y1,'ro--',label='First Line')#设置线条标签
plt.plot(x2,y2,'b-',label='Second Line')
# 设置标题、标签
plt.xlabel('月份')#X轴标签
plt.ylabel('年份')#Y轴标签
plt.title('进出口数据')#标题
#设置X轴范围
plt.xlim(1,3)
#设置Y轴范围
plt.ylim(0,15)
#设置X轴刻度
plt.xticks(np.linspace(0,6,5))
#设置Y轴刻度
plt.yticks(np.arange(1,15,3),['2011年','2012年','2013年','2014年','2015年'])
#获取坐标轴信息
ax=plt.gca()
#设置边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#legend生成默认图例
plt.legend()
plt.show()
```

```python
x1=[1,2,3]
y1=[5,7,4]
x2=[1,2,3]
y2=[10,14,12]
plt.subplot(221)
plt.plot(x1,y1,'r-')
plt.subplot(223)
plt.plot(x2,y2,'b--')
plt.show()
```

```python
#面向对象的写法
fig=plt.figure()
# 在figure上创建图像
ax1=fig.add_subplot(321)
ax2=fig.add_subplot(322)
ax3=fig.add_subplot(312)
ax4=fig.add_subplot(325)
ax5=fig.add_subplot(326)
# 绘图
ax1.plot(np.random.randn(50).cumsum(),'g--')
ax2.plot(np.random.randn(50).cumsum(),'b--')
ax3.plot(np.random.randn(50).cumsum(),'m--')
ax4.plot(np.random.randn(50).cumsum(),'g--')
ax5.plot(np.random.randn(50).cumsum(),'b--')
plt.suptitle('Subplot')
plt.show()
```

```python
Plot,axes=plt.subplots(nrows=4,ncols=1,sharex=True,sharey=False)
Plot.suptitle('test',fontsize=20)# 设置fig标题
axes[0].plot(range(10),'ro-')
axes[1].plot(range(10),'bo-')
axes[2].plot(range(10),'go-')
axes[3].plot(range(10),'mo-')
plt.show()
```

```python
fig,axes=plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(100),10,color='g',alpha=0.75)
fig.subplots_adjust(wspace=1,hspace=1)#调整子图之间的距离
plt.show()
```

```python
plt.figure()
plt.bar([1,3,5,7,9,11],[5,2,7,8,2,6],label="Example one",color='y')
plt.bar([2,4,6,8,10,12],[8,6,2,5,6,3],label="Example two",color='g')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('')
plt.show()
```

```python
n=3 #设置柱子数量
width=0.5/n #设置高度
left=np.arange(1,5)
height=np.array([200,300,400,500])

plt.figure()
plt.bar(left-width,height,width=width,color='g',align='center',label='green')
plt.bar(left,height*3,width=width,color='y',align='center',label='yellow')
plt.bar(left+width,height*2,width=width,color='b',align='center',label='blue')
plt.xticks(np.arange(1,5),[str(i)+'月' for i in np.arange(1,6)])
plt.xlabel('月份')
plt.ylabel('美元')
plt.legend()
plt.show()
```

# 直方图

```python
mu,sigma=100,15
data=np.random.randn(10000)*mu*sigma

# 绘图
plt.figure()
plt.hist(data,100,color='g',alpha=0.75)

plt.grid(True)# 添加网格
plt.show()
```

# 散点图

```python
x=np.random.randn(1000)
y=np.random.randn(1000)
T=np.random.randn(1000)*125
plt.scatter(x,y,label='散点分布',c=T,s=25,marker='o',alpha=0.5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('散点图')
plt.legend()
plt.show()
```

# 饼图

```python
labels=['第一部分','第二部分','第三部分']
explode=(0,0,0.2)
colors=['red','g','y']

# 比例
x=[30,70,20]

plt.figure()
plt.pie(x,explode=explode,colors=colors,labels=labels,startangle=90,shadow=True,
       labeldistance=1.1,autopct='%1.1f%%',pctdistance=0.6)
plt.axis('equal')# 设置饼图为圆的
plt.show()
```
